The threshold of 3 in Z-score-based outlier detection is widely used because, in a normal distribution, it identifies data points that are statistically unusual. According to the empirical rule (68-95-99.7 rule):

- About 68% of data lies within 1 standard deviation of the mean.
- About 95% lies within 2 standard deviations.
- About 99.7% lies within 3 standard deviations.

Using a threshold of 3 means you’re identifying points that fall outside this 99.7% range as potential outliers. However, this threshold isn't rigid; it can be adjusted based on the dataset's characteristics. 

- For more extreme outliers, increase the threshold to 4 or 5.
- For less variability, reduce it to 2 or 2.5.

The goal is to effectively balance identifying true outliers while retaining useful data points.

```python
import matplotlib.pyplot as plt

# Select categorical columns
categorical_cols = data.select_dtypes(include=['object', 'category']).columns
print(categorical_cols)

# Separate the data based on target values
target_0_data = data[data['TARGET'] == 0]
target_1_data = data[data['TARGET'] == 1]

# Loop through each categorical column
for i in categorical_cols:
    plt.figure(figsize=(12, 6))  # Adjust figure size
    
    # Plot for target = 0
    plt.subplot(1, 2, 1)
    target_0_data[i].value_counts(normalize=True).plot(kind='bar', color='skyblue', alpha=0.7)
    plt.title(f'{i} - Target = 0')
    plt.xlabel(i)
    plt.ylabel('Proportion')
    
    # Plot for target = 1
    plt.subplot(1, 2, 2)
    target_1_data[i].value_counts(normalize=True).plot(kind='bar', color='salmon', alpha=0.7)
    plt.title(f'{i} - Target = 1')
    plt.xlabel(i)
    plt.ylabel('Proportion')
    
    plt.tight_layout()
    plt.show() 
```
